Practical 2 : Generate ANDNOT function using McCulloch-Pitts neural net by a python program

```python
import numpy as np
```

1st Method

```python
X=np.array([[0,0],
            [1,0],
            [0,1]])
Y=np.array([1,0,1])
w=np.array([0.0,0.0],dtype=np.float64)
b=0.0
lr=0.1
n_iter=10
def step(x):
    return 1 if x>0 else 0
for i in range(n_iter):
    for j in range(len(X)):
        x=X[j]
        y=Y[j]
        z=np.dot(x,w)+b
        o=step(z)
        dw=lr*(y-o)*x
        db=lr*(y-o)
        w+=dw
        b+=db
test_x=np.array([1,1])
test_z=np.dot(test_x,w)+b
test_o=step(test_z)
print("input:",test_x)
print("Output:",test_o)
```

2nd Method

```python
def mp_neuron(inputs, weights, threshold):
    weighted_sum = np.dot(inputs, weights)
    output = 1 if weighted_sum >= threshold else 0
    return output

def and_not(x1, x2):
    weights = [1, -1]
    threshold = 1
    inputs = np.array([x1, x2])  # Corrected variable name here
    output = mp_neuron(inputs, weights, threshold)
    return output

print(and_not(0, 0))
print(and_not(1, 0))
print(and_not(0, 1))
print(and_not(1, 1))
```
